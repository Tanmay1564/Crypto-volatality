## Value Counts for Categorical Columns

```python
for c in df.select_dtypes(include='object'): print(c, df[c].value_counts(), '\n')
```

## Outlier Detection (IQR)

```python
num_df = df.select_dtypes(include='number')

Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((num_df < (Q1 - 1.5 * IQR)) | (num_df > (Q3 + 1.5 * IQR))).sum()
outliers
```

## Distribution Plots

```python
import matplotlib.pyplot as plt
for col in df.select_dtypes(include='number').columns:
 plt.hist(df[col]); plt.title(col); plt.show()
```

## Boxplots for Numeric Columns

```python
import matplotlib.pyplot as plt
for col in df.select_dtypes(include='number').columns:
 plt.boxplot(df[col]); plt.title(col); plt.show()
```

## Skewness of Numerical Columns

```python
num_df = df.select_dtypes(include='number')

num_df.skew()
```

## Kurtosis of Numerical Columns

```python
num_df = df.select_dtypes(include='number')

num_df.kurt()
```
